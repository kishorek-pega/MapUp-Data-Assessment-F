Question 1: Car Matrix Generation
Under the function named generate_car_matrix write a logic that takes the dataset-1.csv as a DataFrame. Return a new DataFrame that follows the following rules:

values from id_2 as columns
values from id_1 as index
dataframe should have values from car column
diagonal values should be 0.

```python
#initialising the function with a parameter
def generate_car_matrix(dataset_path):


    # Pivot the DataFrame to create a matrix with id_1 as index, id_2 as columns, and car as values
    car_matrix = df.pivot(index='id_1', columns='id_2', values='car').fillna(0)
    
    #returning the pivoted table
    return car_matrix

result_matrix = generate_car_matrix(df)
# printing the table
print(result_matrix)
```

Question 2: Car Type Count Calculation
Create a Python function named get_type_count that takes the dataset-1.csv as a DataFrame. 
Add a new categorical column car_type based on values of the column car:

low for values less than or equal to 15,
medium for values greater than 15 and less than or equal to 25,
high for values greater than 25.
Calculate the count of occurrences for each car_type category and return the result as a dictionary. 
Sort the dictionary alphabetically based on keys.

```python
# creating the function to add a new column and categorize it
def get_type_count():
    
    # creating condition list containing the conditions for the new col
    conditions = [(df['car']<=15), (df['car']>15) & (df['car']<=25), (df['car']>25)]
    
    # taking the categorical value for the above conditions
    value = ['low','medium','high']
    
    #using np.select to map the values to the conditions
    df['car_type'] = np.select(conditions, value)
    
    #printing the dataframe
    print(df)
```

```python
# calling the function
get_type_count()

df
```

```python
# counting the occurances of the values in the car_type col using value_counts()

car_type_count = df['car_type'].value_counts().to_dict()

# sorting the dictoanry into ascending order by key based on the alphabetical order using sorted()
car_type_sort = dict(sorted(car_type_count.items()))

print(car_type_sort)
```

Question 3: Bus Count Index Retrieval
    
Create a Python function named get_bus_indexes that takes the dataset-1.csv as a DataFrame. 

The function should identify and return the indices as a list (sorted in ascending order)
where the bus values are greater than twice the mean value of the bus column in the DataFrame.

```python
#initialing the function 
def get_bus_indexes():
    
    # taking the condition where the bus values are greater than 2*mean
    condition = df[df['bus']>2*(df['bus'].mean())]
    
    # using the sorted function to sort the index of the values returned from the condition
    sorted_index = sorted(condition.index)
    
    #returning the sorted index
    return sorted_index

# calling the function
get_bus_indexes()
```

Question 4: Route Filtering
    
Create a python function filter_routes that takes the dataset-1.csv as a DataFrame. 
The function should return the sorted list of values of column route for 
which the average of values of truck column is greater than 7.

```python
df
```

```python
import pandas as pd

def filter_routes(dataset_path):

    # Group by route and calculate the average of the truck column
    average_truck = df.groupby('route')['truck'].mean()

    # Filter routes where the average of the truck column is greater than 7
    filtered_routes = average_truck[average_truck > 7].index.tolist()

    # Sort the list of routes
    sorted_routes = sorted(filtered_routes)

    return sorted_routes

# Example usage
dataset_path = df 
result = filter_routes(dataset_path)
print(result)
```

Question 6: Time Check
You are given a dataset, dataset-2.csv, containing columns id, id_2, and timestamp 
(startDay, startTime, endDay, endTime). 
The goal is to verify the completeness of the time data by checking whether the timestamps

and span all 7 days of the week (from Monday to Sunday).

Create a function that accepts dataset-2.csv as a DataFrame and returns a boolean series 
that indicates if each (id, id_2) pair has incorrect timestamps. The boolean series must 
have multi-index (id, id_2).

```python
import pandas as pd

def verify_time_data(df: pd.DataFrame) -> pd.Series:
    """
    Verify the completeness of the time data for each (id, id_2) pair.

    Parameters:
        df (pd.DataFrame): The input DataFrame with columns id, id_2, startDay, startTime, endDay, and endTime.

    Returns:
        A boolean series with multi-index (id, id_2) indicating whether the timestamps for each (id, id_2) pair cover a full 24-hour period and span all 7 days of the week.
    """
    # Convert start and end timestamps to datetime format
    df['start'] = pd.to_datetime(df['startDay'] + ' ' + df['startTime'], format='%d %H:%M:%S')
    df['end'] = pd.to_datetime(df['endDay'] + ' ' + df['endTime'], format='%d %H:%M:%S')

    # Calculate the duration of each (id, id_2) pair
    durations = df.groupby(['id', 'id_2'])['start'].diff().dt.total_seconds()

    # Check if the duration is greater than or equal to 86400 (24 hours)
    # and if the start and end timestamps are within the same week
    conditions = [
        durations.gt(86400),  # duration is greater than or equal to 24 hours
        df['start'].dt.weekday.between(0, 6)  # start and end timestamps are within the same week
    ]

    # Create a boolean series with multi-index (id, id_2)
    result = pd.Series(conditions, index=df.index, name='valid')

    return result

df = pd.read_csv('dataset-2.csv')
result = verify_time_data(df)
```
